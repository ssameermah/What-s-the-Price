```python
imgsrc = []


def get_url(url):
    userAgent = {'User-agent': 'Mozilla/5.0'}
    response = requests.get(url, headers=userAgent)
    req=requests.get(url,headers=userAgent)
    content=req.text
    print("Request made")

    soup=BeautifulSoup(content)
    x = soup.find_all('div',{'class':'article-image-container'})
    for i in range(len(x)):
        a = str(x[i].find('div',{'class':'content'}).find('img'))
        if 'data-original=' and 'data-srcset=' in a:
            b = a.split('data-original=')[1].split('data-srcset=')[0]
            imgsrc.append(b)
            print(b)
        elif 'src=' and 'srcset=' in a:
            b = a.split('src=')[1].split('srcset=')[0]
            imgsrc.append(b)
            print(b)
        else:
            print(f'{i} is weird')
    
brands = ['rolex','casio','breitling','gucci','seiko']
L=[]
for i in range(1,50):
    for j in brands:
        url = f'https://www.chrono24.com/{j}/index-{i}.htm?query={j}'
        get_url(url)
print(len(imgsrc))  
```

```python
title = []

def get_name(url):
    userAgent = {'User-agent': 'Mozilla/5.0'}
    response = requests.get(url, headers=userAgent)
    req=requests.get(url,headers=userAgent)
    content=req.text
    print("Request made")

    soup=BeautifulSoup(content)
    
    y = soup.find_all('div',{'class':'article-title'})

    for i in range(len(y)):
        a = str(y[i])
        a = a.split('\n         ')[1].split('<')[0]
        title.append(a)
        print(a)
    
brands = ['rolex','casio','breitling','gucci','seiko']
L=[]
for i in range(1,50):
    for j in brands:
        url = f'https://www.chrono24.com/{j}/index-{i}.htm?query={j}'
        get_name(url)
print(len(title)) 
```

```python
price = []
def get_price(url):
    userAgent = {'User-agent': 'Mozilla/5.0'}
    response = requests.get(url, headers=userAgent)
    req=requests.get(url,headers=userAgent)
    content=req.text
    print("Request made")

    soup=BeautifulSoup(content)
    

    z = soup.find_all('div',{'class':'article-price'})

    for i in range(len(z)):
        a = str(z[i])
        if '</span>' in a:
            a = a.split('</span>')[1].split('</strong>')[0]
            price.append(a)
            print(a)
        elif 'Price on request' in a:
            a = a.split('-information">')[1].split('</strong>')[0]
            price.append(a)
            print(a)
    
brands = ['rolex','casio','breitling','gucci','seiko']
# remove armani, fossil
L=[]
for i in range(1,50):
    for j in brands:
        url = f'https://www.chrono24.com/{j}/index-{i}.htm?query={j}'
        get_price(url)
print(len(price))
```

```python
from pathlib import Path
import tensorflow as tf
image_dir = Path('/Users/sameermahajan/Downloads/output/train')
```

```python
import os

num_skipped = 0
for folder_name in ('Rolex','Casio','Breitling','Gucci','Seiko'):
    folder_path = os.path.join(image_dir, folder_name)
    for fname in os.listdir(folder_path):
        fpath = os.path.join(folder_path, fname)
        try:
            fobj = open(fpath, "rb")
            is_jfif = tf.compat.as_bytes("JFIF") in fobj.peek(10)
        finally:
            fobj.close()

        if not is_jfif:
            num_skipped += 1
            # Delete corrupted image
            os.remove(fpath)

print("Deleted %d images" % num_skipped)
```
